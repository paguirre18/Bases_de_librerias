**1. Carga y exploración del dataset**

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
df.head()
```

```python
df.info()
```

El dataset tiene 5 columnas: 4 de dimensiones de la planta y 1 del tipo de especie. La estructura del dataset permite plantear un problema de clasificación: predecir la especie de la planta según sus dimensiones

```python
df["species"].unique()
```

**2. Gráficos con Matplotlib**

```python
petal_proms = df.groupby("species")[["petal_length", "petal_width"]].mean()

petal_proms.plot(kind="bar")
plt.title("Promedio de largo y ancho de pétalos por especie")
plt.ylabel("Promedio (cm)")
plt.xlabel("Especie")
plt.show()
```

```python
plt.hist(df['sepal_length'], bins=18, color='green')
plt.title("Histograma de largo de sépalos")
plt.xlabel("Largo de sépalos (cm)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df, palette="rainbow")
plt.title("Largo vs Ancho de sépalos por especie")
plt.xlim(3, 8)
plt.ylim(1, 5)
plt.xlabel("Largo de sépalos (cm)")
plt.ylabel("Ancho de sépalos (cm)")
plt.legend(title="Especie")
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_length", data=df, color="lightblue")
plt.title("Largo de sépalo por especie")
plt.xlabel("Especie")
plt.ylabel("Largo (cm)")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.boxplot(x='species', y='sepal_length', data=df, ax=axes[0], palette='Set2')
axes[0].set_title('Largo de sépalo')

# Sepal Width
sns.boxplot(x='species', y='sepal_width', data=df, ax=axes[1], palette='Set3')
axes[1].set_title('Sepal Width by Species')

plt.tight_layout()
plt.show()
```

```python
plt.plot(df.index, df['petal_length'], label='Largo de sépalos', color='blue')
plt.title('Largo de sépalos a lo largo de las observaciones')
plt.xlabel("Observación")
plt.ylabel("Largo de pétalos (cm)")
plt.show()
```
