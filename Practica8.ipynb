**1. Carga y exploración del dataset**

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Se trabja con datos categóricos y numéricos. No hay datos faltantes

**2. Gráficos con Seaborn**

```python
#Es una matriz de correlación
data = df.loc[:, ['total_bill', 'tip', 'size']].corr()

sns.heatmap(data, cmap="coolwarm", annot=True)
plt.show()
```

```python
sns.violinplot(x="day", y="tip", data=df)
plt.title("Distribución de propinas por día")
plt.xlabel("Día")
plt.ylabel("Propina ($)")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df, hue="day", palette="rainbow")
plt.title("Total de cuenta vs Propina")
plt.xlabel("Cuenta ($)")
plt.ylabel("Propina ($)")
plt.legend(title="Día")
plt.show()
```

```python
sns.barplot(x="sex", y=df["tip"].mean(), data=df, color="lightgreen")
plt.title("Promedio de propinas")
plt.xlabel("Sexo")
plt.ylabel("Propina ($)")
plt.show()
```
